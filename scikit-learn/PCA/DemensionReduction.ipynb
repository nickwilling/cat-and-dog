```python
"""找出n个新的特征向量,让数据能够被压缩到少数特征上并且总信息量不损失太多的技术
就是矩阵分解,PCA和SVD是两种不同的降维算法,但他们都是从上面的过程来实现降维,只是
两种算法中矩阵分解的方法不同,信息量的衡量指标不同罢了 """
#PCA使用特征值分解来找出空间V使用方差来作为信息量的衡量指标,方差大的特征所含的信息量比较多会被保留下来
#SVD使用奇异值分解来找出空间V使用奇异值来衡量特征上的信息量
```

```python
#高维数据可视化,可视化一般是降到3维或者2维看数据的分布
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
X = iris.data
y = iris.target
import pandas as pd
pd.DataFrame(X)
```

```python
#调用PCA
pca = PCA(n_components=2)#实例化
pca = pca.fit(X)#拟合模型
X_dr = pca.transform(X)#获取降维后的矩阵,就变成二维的 了
X_dr
```

```python
#可视化:目标,label有3种花,要呈现三种花在坐标轴的分布
plt.figure
print(X_dr.shape)
print(y.shape)
plt.scatter(X_dr[:,0],X_dr[:,1],c=y)
plt.show()
```

```python

y==0#y矩阵里面有012,y==0会返回每个数据是不是等于0,True/False
```

```python
#使用布尔索引
#不加这一行图像就不显示在笔记本上而是另开一个窗口了
%matplotlib inline
plt.figure()
plt.scatter(X_dr[y==0,0],X_dr[y==0,1],c="red",label=iris.target_names[0])#返回y==0的那些列的第一个特征和第二个特征
plt.scatter(X_dr[y==1,0],X_dr[y==1,1],c="k",label=iris.target_names[1])
plt.scatter(X_dr[y==2,0],X_dr[y==2,1],c="orange",label=iris.target_names[2])
plt.legend()
plt.title("PCA od IRIS dataset")
plt.show()
```

```python
#属性explained_variance_，查看降维后每个特征向量上所带的信息量的大小（可解释性方差的大小）
pca.explained_variance_
```

```python
#属性explained_variance_ratio,查看降维后每个新特征向量所占的信息量占原始数据总信息量的百分百
#又叫可解释方差贡献率
pca.explained_variance_ratio_
```

```python
#大部分信息都被有效的集中在了第一个特征上
pca.explained_variance_ratio_.sum()
```

```python
PCA().fit(X).explained_variance_ratio_
```

```python
#选择最好的n_components:累积可解释方差贡献率曲线
#降维后保留的特征个数为横坐标，降维后新特征矩阵捕捉到的可解释方差贡献率为纵坐标
import numpy as np
pca_line = PCA().fit(X)
plt.plot([1,2,3,4],np.cumsum(pca_line.explained_variance_ratio_))
plt.xticks([1,2,3,4])
plt.xlabel("number of components after dimension reduction")
plt.ylabel("cumulative explained variance")
plt.show()
```

```python
#最大似然估计自选超参数
pca_mle = PCA(n_components="mle")
pca_mle = pca_mle.fit(X)
X_mle = pca_mle.transform(X)
X_mle#可以发现，mle为我们自动选择了三个特征
pca_mle.explained_variance_ratio_.sum()
```

```python
#按信息量占比选超参数
pca_f = PCA(n_components=0.97,svd_solver="full")#输入0.97，PCA会自动选出保留信息量超过97%的特征数量
pca_f = pca_f.fit(X)
x_f = pca_f.transform(X)
pca_f.explained_variance_ratio_
```

```python
import logging
from time import time
from numpy.random import RandomState
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
from sklearn.cluster import MiniBatchKMeans
from sklearn import decomposition
```

```python
# Display progress logs on stdout
logging.basicConfig(level=logging.INFO,
format='%(asctime)s %(levelname)s %(message)s')
n_row, n_col = 2, 3
n_components = n_row * n_col
image_shape = (64, 64)
rng = RandomState(0)
```

```python
# #############################################################################
# Load faces data
dataset = fetch_olivetti_faces(shuffle=True, random_state=rng)
faces = dataset.data
```

```python
n_samples, n_features = faces.shape
```

```python
# global centering
faces_centered = faces - faces.mean(axis=0)
```

```python
# local centering
faces_centered -= faces_centered.mean(axis=1).reshape(n_samples, -1)
```

```python
print("Dataset consists of %d faces" % n_samples)
```

```python
def plot_gallery(title, images, n_col=n_col, n_row=n_row, cmap=plt.cm.gray):
    plt.figure(figsize=(2. * n_col, 2.26 * n_row))
    plt.suptitle(title, size=16)
    for i, comp in enumerate(images):
        plt.subplot(n_row, n_col, i + 1)
        vmax = max(comp.max(), -comp.min())
        plt.imshow(comp.reshape(image_shape), cmap=cmap,
        interpolation='nearest',
        vmin=-vmax, vmax=vmax)
        plt.xticks(())
        plt.yticks(())
    plt.subplots_adjust(0.01, 0.05, 0.99, 0.93, 0.04, 0.)
```

```python
estimators = [
('Eigenfaces - PCA using randomized SVD',
decomposition.PCA(n_components=n_components, svd_solver='randomized',
whiten=True),
True),
('Non-negative components - NMF',
decomposition.NMF(n_components=n_components, init='nndsvda', tol=5e-3),
False),
('Independent components - FastICA',
decomposition.FastICA(n_components=n_components, whiten=True),
True),
('Sparse comp. - MiniBatchSparsePCA',
decomposition.MiniBatchSparsePCA(n_components=n_components, alpha=0.8,
n_iter=100, batch_size=3,
random_state=rng,
normalize_components=True),
True),
('MiniBatchDictionaryLearning',
decomposition.MiniBatchDictionaryLearning(n_components=15, alpha=0.1,
n_iter=50, batch_size=3,
random_state=rng),
True),
('Cluster centers - MiniBatchKMeans',
MiniBatchKMeans(n_clusters=n_components, tol=1e-3, batch_size=20,
max_iter=50, random_state=rng),
True),
('Factor Analysis components - FA',
decomposition.FactorAnalysis(n_components=n_components, max_iter=20),
True),
]
```

```python
plot_gallery("First centered Olivetti faces", faces_centered[:n_components])
```

```python
for name, estimator, center in estimators:
    print("Extracting the top %d %s..." % (n_components, name))
    t0 = time()
    data = faces
    if center:
        data = faces_centered
    estimator.fit(data)
    train_time = (time() - t0)
    print("done in %0.3fs" % train_time)
    if hasattr(estimator, 'cluster_centers_'):
        components_ = estimator.cluster_centers_
    else:
        components_ = estimator.components_
        
    if (hasattr(estimator, 'noise_variance_') and
            estimator.noise_variance_.ndim > 0): # Skip the Eigenfaces case
        plot_gallery("Pixelwise variance",
            estimator.noise_variance_.reshape(1, -1), n_col=1,
            n_row=1)
    plot_gallery('%s - Train time %.1fs' % (name, train_time),
            components_[:n_components])
plt.show()
```

```python
estimators = [
('Dictionary learning',
decomposition.MiniBatchDictionaryLearning(n_components=15, alpha=0.1,
n_iter=50, batch_size=3,
random_state=rng),
True),
('Dictionary learning - positive dictionary',
decomposition.MiniBatchDictionaryLearning(n_components=15, alpha=0.1,
n_iter=50, batch_size=3,
random_state=rng,
positive_dict=True),
True),
('Dictionary learning - positive code',
decomposition.MiniBatchDictionaryLearning(n_components=15, alpha=0.1,
n_iter=50, batch_size=3,
random_state=rng,
positive_code=True),
True),
('Dictionary learning - positive dictionary & code',
decomposition.MiniBatchDictionaryLearning(n_components=15, alpha=0.1,
n_iter=50, batch_size=3,
random_state=rng,
positive_dict=True,
positive_code=True),
True),
]
```

```python
plot_gallery("First centered Olivetti faces", faces_centered[:n_components],
cmap=plt.cm.RdBu)
```

```python
for name, estimator, center in estimators:
    print("Extracting the top %d %s..." % (n_components, name))
    t0 = time()
    data = faces
    if center:
        data = faces_centered
    estimator.fit(data)
    train_time = (time() - t0)
    print("done in %0.3fs" % train_time)
    components_ = estimator.components_
    plot_gallery(name, components_[:n_components], cmap=plt.cm.RdBu)

plt.show()
```
