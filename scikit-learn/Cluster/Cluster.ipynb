```python
#聚类前先可视化一下，看看究竟有多少个类
#自己创建数据集
from sklearn.datasets import make_blobs
X,y = make_blobs(n_samples=500,n_features=2,centers=4,random_state=1)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig,ax1 = plt.subplots(1)
ax1.scatter(X[:,0],X[:,1],marker='o',s=8,c=y)
plt.legend()
plt.show()
```

```python
#如果我们想要看见这个点的分布，怎么办？
color = ["red","pink","orange","gray"]
fig, ax1 = plt.subplots(1)
for i in range(4):
    ax1.scatter(X[y==i,0],X[y==i,1],c=color[i],marker='x')
plt.show()
```

```python
from sklearn.cluster import KMeans

n_clusters = 3

cluster = KMeans(n_clusters=n_clusters,random_state=0).fit(X)
```

```python
y_pred = cluster.labels_
```

```python
#重要属性簇心
centroid =cluster.cluster_centers_
centroid
```

```python
centroid.shape
```

```python
#重要属性inertia_，查看总距离的平方和
inertia = cluster.inertia_
inertia
```

```python
color = ["red","pink","orange","gray"]
fig, ax1 = plt.subplots(1)
for i in range(n_clusters):
    ax1.scatter(X[y_pred==i,0],X[y_pred==i,1],s=8,marker="o",c=color[i])

ax1.scatter(centroid[:,0],centroid[:,1],marker="x",s=35,c='black')

plt.show()
```

```python
#聚类算法的模型评估指标
#Kmeans的目标是确保“簇内差异小，粗外差异大”，我们就可以通过衡量簇内差异来衡量聚类的效果
#K_means是对数据有假设的，inertia容易受到数据的影响，就是假设数据是一簇一簇的，对于细长簇，环形簇或者不规则形状的流形时表现不佳。
#标签未知时：轮廓系数：可以用来评价簇内的稠密程度（簇内差异小）和簇间的离散程度（簇外差距大）来评估聚类的效果
#轮廓系数的计算：s=(b-a)/max(a,b),a为簇内，b为簇间【-1，1】
from sklearn.metrics import silhouette_score
from sklearn.metrics import silhouette_samples
from sklearn.metrics import calinski_harabaz_score
```

```python
print(silhouette_score(X,cluster_.labels_))
print(calinski_harabaz_score(X,cluster_.labels_))
```

```python
silhouette_samples(X,cluster_.labels_) > 0
```
